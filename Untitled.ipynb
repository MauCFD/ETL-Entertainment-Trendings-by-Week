```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = "https://www.boxofficemojo.com/weekly/by-year/"
complementary_url = "/?sortDir=asc&sort=weekNum&ref_=bo_wly__resort#table"
```

```python
movies_dict = []

for x in range(1999,2020):
        url = base_url+str(x)+complementary_url
        
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        
        movies = soup.find_all('td', class_ = 'a-text-left mojo-field-type-release mojo-cell-wide')
        movies[0].select('a')[0].string

        week_number = soup.find_all('td', class_ = 'a-text-right mojo-field-type-date_interval mojo-sort-column')
        week_number[0].select('a')[0].string
        
        for y, z in zip(movies, week_number):
                movies_dict.append({"Year": x, 
                                    "Week": z.select('a')[0].string,
                                    "Movie": y.select('a')[0].string})
            
movies_dict
```

```python
movie_df = pd.DataFrame(movies_dict)
movie_df
```
